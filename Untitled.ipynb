```python
print(horse2.groupby('outcome')['pain'].median())
fill_median_by_phy = horse2.copy()
fill_median_by_phy['pain'] = fill_median_by_phy['pain'].fillna(horse2.groupby('outcome')['pain'].transform('median'))
#при летальном исходе (собственная смерть) боль лошади сильнее по субьетивной оценке.
```

```python
fill_median_by_phy.head(50)
```

```python
#Теперь рассчитаем значения по пропускам из некатегориальных значений 'pulse_ok','respiratory_rate_ok','rectal_temperature_ok' исходя из уровня боли лошади. 
fill_median_by_pain = fill_median_by_phy.copy()
fill_median_by_pain[['pulse_ok','respiratory_rate_ok','rectal_temperature_ok']] = fill_median_by_pain[['pulse_ok','respiratory_rate_ok','rectal_temperature_ok']].fillna(horse2.groupby('pain')[['pulse_ok','respiratory_rate_ok','rectal_temperature_ok']].transform('median'))
```

```python
fill_median_by_pain.info()
```

```python
#немногие оставшиеся пропуски заполним модой, которая минимально исказит данные
fill_by_mode = fill_median_by_pain.copy()
fill_by_mode['pulse_ok'] = fill_by_mode['pulse_ok'].fillna(fill_median_by_pain['pulse_ok'].mode()[0])
fill_by_mode['respiratory_rate_ok'] = fill_by_mode['respiratory_rate_ok'].fillna(fill_median_by_pain['respiratory_rate_ok'].mode()[0])
fill_by_mode['rectal_temperature_ok'] = fill_by_mode['rectal_temperature_ok'].fillna(fill_median_by_pain['rectal_temperature_ok'].mode()[0])
```
