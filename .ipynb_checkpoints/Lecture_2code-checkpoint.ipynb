```python
#let's start with the simplest
#possible code
y = "Hello World"
print(y)
#y is a variable that stores a string
#print function prints y in the console
#on the right
#variables can be many things
y = 1
print(y)
#often you'll need to use a specific library
#you can load the whole library
#with import command
import pandas as pd
#as operator allows you to shorten the key
#word for the package 
#you can also import specific functions
#and methods from each library
#let's work with string for a bit
#first, we can access a specific element
#of a string by selecting its index
y = "PoliSci"
print(y[0])
#note that Python is 0-indexed language!
#you can also index the last element
print(y[-1])
#and the last but one
print(y[-2])
#note that there is no -0 indexing!
#finally, we can slice strings by indices
print(y[0:2])
#we can even do reverse slicing!
print(y[::-1])
print(y[3::-1])
#strings can be concatenated with +
y = y + " NewStr"
print(y)
#we can store multiple elements
#in Python lists
y = [1, "string", 2,0]
print(y)
#as you can see, string element can
#be absolutely anything, as long 
#as its supported by Python
#nested lists are also valid
c = [1, "string", y, 2,0]
print(c)
#you can access element of a list 
#in the same way as string
print(y[0])
#for nested lists, you need to use double indexing
print(c[2][1])
#lists, unlike strings, can be mutated with sub-indexing:
c[0] = 5
print(c)
#you can add elements to lists using append
c.append("New El!")
print(c)
#you can also use +
c = c + ["New El 2!"]
#notice [] brackets - only a list can be added to a list!
#also, please, note that operations c + ["New El 2!"] 
#and ["New El 2!"] + c are not computationally equivalent!
#dictionaries constitute another important data structure
#that Python supports
#dictionaries are collections of named items
dict_example = {"name": "Evgeny", "surname": "Sedashov"}
#the elements to the left of : sign are called keys
#keys cannot be mutable objects, eliminating lists as keys
#normally, you would want strings or integers as keys
#dictionaries are themselves mutable
dict_example["name"] = "John"
print(dict_example)
#you can see that this works
#finally, there are pandas data frames
data_frame = pd.DataFrame({"GDP_PC": [10743, 75505, 53442, 15594], 
                           "Countries": ["Russia", "Norway", "Sweden", "Latvia"]})
print(data_frame)
#you can create new variable in a dataframe
data_frame["New Var"] = [4,3,0,-2]
data_frame
#you can also subset dataframe by conditions
data_frame.loc[data_frame.Countries == "Russia", ["GDP_PC", "New Var"]]
data_frame.loc[data_frame.Countries == "Russia"]
#functions allow to automate certain operations on the same
#type of data
def print_name(my_name):
    print("My name is " + my_name)
print_name("John")
#function definitions ALWAYS start with the word def, then
#goes the function name, and then inputs; default inputs ARE allowed
def print_name(my_name = "Evgeny"):
    print("My name is " + my_name)
print_name()
print_name("Carl")
#general advice is to simply Google how to do a certain
#operation; there is not a single programming question I had in my experience
#that I could not answer by simple Google search 

#control flow 
#in programming if statements and loops are of crucial importance
#here is the basic syntax for if statement
x = 2
if x<3:
    print("x is small")
else:
    print("x is large")
#more complex conditioning is also possible
x = 5
if x<3:
    print("x is small")
elif x>=3 and x<=5:
    print("x is midsize")
else:
    print("x is large")

#loops is one of the most important concept in programming
#the idea is of repeting a certain sequence of operations
#some pre-specified number of times
x = 0
while x < 100:
    print(x)
    x += 1
#here, we print x and then increase it by adding one unit 100 times
#equivalently, we could have written print(x) x+=1 100 times
#but that would have cost us time, not to mention routine nature
#of the task... 
#almost every advanced algorithm includes iterations
#so the importance of learning how to do it cannot be overstated
#using for loops is often simpler
for x in range(100):
    print(x)
#range() function generates our 100 x
#we can use for loops to loop through any iterable such as lists
#let's use looping techniques to write a simple function
def max_val(vector = [0]):
    max_value = vector[0]
    #note how we slice the input vector to exclude the first element
    for x in vector[1:]:
        if x>=max_value:
            max_value = x
    return max_value
print(max_val())
print(max_val([3,1,5,-2]))
```

```python
#loops can also be nested
for x in (range(100)):
    for y in range(10):
        print(str(x) + str(y))
```
