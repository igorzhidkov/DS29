# **Task 1**

```python
n_df = pd.read_csv('visit_log.csv', sep=';')
n_df.head()
```

```python
def sorted_source(row):
    '''
       если источник traffic_source равен yandex или google, то в source_type ставится organic
       для источников paid и email из России - ставим ad
       для источников paid и email не из России - ставим other
       все остальные варианты берем из traffic_source без изменений
       '''
    
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif row['region'] == 'Russia' and row['traffic_source'] == 'paid' or row['region'] == 'Russia'and row['traffic_source'] == 'email':
        return 'ad'
    elif row['region'] != 'Russia':
        return 'other'
    
       
```

```python
n_df['source_type'] = n_df.apply(sorted_source, axis=1)
n_df.head()
```

```python
n_df[n_df['source_type'] == 'ad' ].head()
```

# **Task 3**

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.head()

# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается
# разница между максимальным и минимальным значением столбца timestamp для данного значения userId.
```

```python
more100 = ratings['userId'].value_counts()
more100.head()
```

```python
df = pd.DataFrame(more100)
df1 = df.reset_index()
```

```python
ndf = df1[ df1.userId > 100 ].rename(columns={'index': 'userId', 'userId': 'rate_count'})
ndf.tail()
# df = df.
```

```python
new_ratings = ratings.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
new_ratings.head()
```

```python
new_ratings['livetime'] = new_ratings['max'] - new_ratings['min']
```

```python
def funktion(row):
    return row['max'] - row['min']
```

```python
new_ratings['livetime'] = new_ratings.apply(funktion, axis=1)
new_ratings.head()
```

```python
joined = ndf.merge(new_ratings, on='userId', how='left')
del joined['max']
del joined['min']
del joined ['rate_count'] 
joined.tail()
```
